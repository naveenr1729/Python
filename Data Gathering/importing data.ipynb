CSV:</h4>A CSV (Comma-Separated Values) file is a plain text file format used to store tabular data, such as a spreadsheet or a database. # Convertors

```python
def rename(name):
    if name=="Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
df=pd.read_csv("D:\Python\Datasets\ipl-matches.csv",converters={"Team1":rename})
df.head(3)
```
