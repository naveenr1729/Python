```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2-D Plot
* Used for Bivariate analysis.
* Used for Categorial VS. Numerical.
* used for Numerical VS. Numerical as well.

```python
d1=pd.read_csv("D:\VS code\Datasets\sharma-kohli.csv")
d1
```

```python
# plotting single graph
plt.plot(d1["index"],d1["V Kohli"])
```

```python
# plotting multiple graphs
plt.plot(d1["index"],d1["V Kohli"])
plt.plot(d1["index"],d1["RG Sharma"])
```

```python
# Graph labelling
plt.plot(d1["index"],d1["V Kohli"])
plt.plot(d1["index"],d1["RG Sharma"])
plt.title("Rohit Sharma VS. Virat Kohli ipl runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
# Colors(hex) and line(width and style) and markers
plt.plot(d1["index"],d1["V Kohli"],color="#39CCD6")
plt.plot(d1["index"],d1["RG Sharma"],color="#79210E")
plt.title("Rohit Sharma VS. Virat Kohli ipl runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(d1["index"],d1["RG Sharma"],color="red",linestyle="solid",linewidth="4",marker="8",markersize="12")
plt.plot(d1["index"],d1["V Kohli"],color="#39CCD6",linestyle="dashed",linewidth="3",marker="o") 
```

```python
# Adding legend
plt.plot(d1["index"],d1["RG Sharma"],color="#CCD639",linestyle="solid",linewidth="3",marker="8",markersize="7",label="RG Sharma")
plt.plot(d1["index"],d1["V Kohli"],color="#39CCD6",linestyle="dotted",linewidth="3",marker="o",label="V Kohli") 
plt.legend(loc="best")
```

# Scatter Plots
* Used for Bivariate Analysis
* Numerical VS. Numerical
* Use case - Correlational Analysis

```python
x=np.linspace(-10,10,50)
y=10*x+3+np.random.randint(0,300,50)

# To plot scatterplot
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("x vs. y")
```

# Bar chart
* Used for Bivariate Analysis.
* Numerical VS. Categorical.
* Aggregate values are taken in y-axis(Numerical) 

```python
# Simple bar chat
children=[10,20,40,10,30]
colors=["red","blue","green","yello","pink"]
```

```python
plt.bar(colors,children)
plt.xlabel("colors")
plt.ylabel("children")
```

# Horizontal bar chart
* Used when number of categories are more

```python
plt.barh(colors,children)
```
