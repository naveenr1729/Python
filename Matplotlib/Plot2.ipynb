# Multiple bar plots

```python
d1=pd.read_csv(r"D:\VS code\Datasets\batsman_season_record.csv")
d1
```

```python
plt.bar(d1["batsman"],d1["2015"])
plt.bar(d1["batsman"],d1["2016"])
plt.bar(d1["batsman"],d1["2017"])
```

```python
np.arange(d1.shape[0])
```

```python
plt.bar(np.arange(d1.shape[0])-0.2,d1["2015"],width=0.2,color="yellow")
plt.bar(np.arange(d1.shape[0]),d1["2016"],width=0.2)
plt.bar(np.arange(d1.shape[0])+0.2,d1["2017"],width=0.2)

plt.xticks(np.arange(d1.shape[0]),d1["batsman"])
plt.show()
```

# If the x-axis names are overlapping

```python
children=[10,20,40,10,30]
colors=["red red red red red red","blue blue blue blue blue blue","green green green green green green","yellow yellow yellow yellow yellow yellow","pink pink pink pink pink pink"]
```

```python
plt.bar(colors,children)
plt.xticks(rotation="vertical")
plt.show()
```

# Stacked bar chart

```python
plt.bar(d1["batsman"],d1["2017"],label="2017")
plt.bar(d1["batsman"],d1["2016"],bottom=d1["2017"],label="2016")
plt.bar(d1["batsman"],d1["2015"],bottom=(d1["2016"]+d1["2017"]),label="2016")

plt.legend()
plt.show()
```

# Pie Chart
* Univariate/Bivariate Analysis
* Categorical VS. Numerical
* Use Case- To find contribution on a standard scale

```python
data=[20,45,100,28,45]
subjects=["english","kannnada","science","maths","social"]
plt.pie(data,labels=subjects)
plt.show()
```

```python
d3=pd.read_csv(r"D:\VS code\Datasets\gayle-175.csv")
d3
```

```python
plt.pie(d3["batsman_runs"],labels=d3["batsman"],autopct="%0.1f%%")
plt.show()
```

# To get percentage representation

```python
plt.pie(d3["batsman_runs"],labels=d3["batsman"],autopct="%0.1f%%",colors=["blue","red","green","yellow","grey","black"])
plt.show()
```

# To explode certain slice

```python
plt.pie(d3["batsman_runs"],labels=d3["batsman"],autopct="%0.1f%%",colors=["blue","red","green","yellow","grey","black"],explode=[0.1,0,0,0,0,0.1],shadow=True)
plt.show()
```
