```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1=pd.read_csv(r"D:\VS code\Datasets\iris.csv")
d1.sample(5)
```

# Scatterplot

```python
d1["Species"]=d1["Species"].replace({'Iris-setosa':0,'Iris-versicolor':1,"Iris-virginica":2})
d1.sample(5)
```

```python
plt.scatter(d1["SepalLengthCm"],d1["PetalLengthCm"],c=d1["Species"],cmap="jet",alpha=.7)
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
plt.colorbar()
```

```python
plt.scatter(d1["PetalLengthCm"],d1["PetalWidthCm"],c=d1["Species"])
plt.colorbar()
```

# Change the plot size

```python
plt.figure(figsize=(10,7))

plt.scatter(d1["SepalLengthCm"],d1["PetalLengthCm"],c=d1["Species"],cmap="jet",alpha=1)
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
plt.colorbar()

plt.savefig("figsize")
```

# Annotations

```python
d2=pd.read_csv(r"D:\VS code\Datasets\batter.csv")
d2
```

```python
sample_d2=d2.head(100).sample(25,random_state=5)
sample_d2
```

```python
plt.figure(figsize=(15,7))

plt.scatter(sample_d2["avg"],sample_d2["strike_rate"],s=sample_d2["runs"])
for i in range(sample_d2.shape[0]):
    plt.text(sample_d2["avg"].values[i],sample_d2["strike_rate"].values[i],sample_d2["batter"].values[i])
```

# To create horizontal and vertical lines


```python
plt.scatter(sample_d2["avg"],sample_d2["strike_rate"],s=sample_d2["runs"])
plt.axhline(120,color="red")
plt.axhline(140,color="green")
plt.axvline(35,color="red")

for i in range(sample_d2.shape[0]):
    plt.text(sample_d2["avg"].values[i],sample_d2["strike_rate"].values[i],sample_d2["batter"].values[i])
```

# Subplots

```python
d2.head()
```

```python
plt.scatter(d2["avg"],d2["strike_rate"])
```

```python
fig,ax = plt.subplots()

ax.scatter(d2["avg"],d2["strike_rate"],color="#003399",marker="+")
ax.set_title("Subplot 1")
ax.set_xlabel("Average")
ax.set_ylabel("Strike Rate")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].scatter(d2["avg"],d2["strike_rate"])
ax[1].scatter(d2["avg"],d2["runs"])

ax[0].set_title("Avg Vs Strike Rate")
ax[0].set_ylabel("Strike Rate")
ax[0].set_xlabel("Average")

ax[1].set_title("Avg Vs Runs")
ax[1].set_ylabel("Runs")
ax[1].set_xlabel("Average")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,6))
ax[0,0].scatter(d2["avg"],d2["strike_rate"])
ax[0,1].scatter(d2["avg"],d2["runs"])

ax[1,0].hist(d2["strike_rate"])
ax[1,1].hist(d2["runs"])
```

# Different Method

```python
fig=plt.figure(figsize=(11,7))

ax1=fig.add_subplot(2,2,1)
plt.tight_layout() # avoid overlapping
ax1.scatter(d2["avg"],d2["strike_rate"])
ax1.set_title("Avg VS. Strike Rate")

ax2=fig.add_subplot(2,2,2)
ax2.scatter(d2["avg"],d2["runs"])
ax2.set_title("Avg VS. Runs")

ax3=fig.add_subplot(2,2,3)
ax3.hist(d2["strike_rate"])
ax3.set_title("Histogram of Strike Rate")

ax4=fig.add_subplot(2,2,4)
ax4.hist(d2["runs"])
ax4.set_title("Histogram of Runs")
```
