```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 3D Scatter plot

```python
d1=pd.read_csv(r"D:\VS code\Datasets\batter.csv")
d1
```

```python
fig=plt.figure(figsize=(15,8))
ax=plt.subplot(projection="3d")

ax.scatter3D(d1["runs"],d1["avg"],d1["strike_rate"])
ax.set_title("Performance Analysis")
ax.set_xlabel("Runs")
ax.set_ylabel("Avg")
ax.set_zlabel("Strike Rate")

```

# 3d Line plot

```python
x=[0,1,5,24]
y=[0,10,13,75]
z=[0,13,20,43]

fig=plt.figure()
ax=plt.subplot(projection="3d")

ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z)
```

# 3D Surface Plots

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,6))
ax=plt.subplot(projection="3d")

p=ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
```

```python
fig=plt.figure(figsize=(12,6))
ax=plt.subplot(projection="3d")

p=ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

# Contour Plot
* Converts 3D surface plot into 2D.

```python
z=xx**2+yy**2
```

```python
fig=plt.figure(figsize=(9,5))
ax=plt.subplot()

p=ax.contour(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
fig=plt.figure()
ax=plt.subplot()

p=ax.contourf(xx,yy,z)
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
```

```python
fig=plt.figure()
ax=plt.subplot()

p=ax.contourf(xx,yy,z)
fig.colorbar(p)
```
