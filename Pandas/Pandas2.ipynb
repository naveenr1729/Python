# Rank(series)

```python
d3=pd.read_csv("Datasets/batsman_runs_ipl.csv")
d3
```

```python
d3.sample(5)
```

```python
d3["batsman_rank"]=d3['batsman_run'].rank(ascending=False)
d3
```

```python
d3.sort_values("batsman_rank")
```

# Sort index

```python
d3.sort_index(ascending=False)
```
