```python
import pandas as pd
import numpy as np
```

Pandas Series: A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

# To create series from lists:

```python
#Strings
country=["India","Pakistan","USA","Nepal"]
pd.Series(country)
```

```python
runs=[25,56,45,65,32,54,89]
pd.Series(runs)
```

Custom index

```python
marks=[64,56,45,100]
subjects=["maths","science","english","hindi"]
pd.Series(marks,index=subjects)
```

```python
marks=[64,56,45,100]
subjects=["maths","science","english","hindi"]
pd.Series(marks,index=subjects,name="Marks of Naveen")
```

# To create series from dictionary

```python
marks={"maths":85,"english":79,"science":86,"social":45}
marks_series=pd.Series(marks,name="Marks of Naveen")
marks_series
```
