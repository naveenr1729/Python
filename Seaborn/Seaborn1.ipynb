# Scatterplot

```python
d1=pd.read_csv(r"D:\Python\Datasets\tips.csv")
d1
```

```python
# scatterplot is an axis level function
sns.scatterplot(data=d1,x="total_bill",y="tip",hue="sex",style="time")
```

* **hue:** You can use the hue parameter to color the points in the plot according to a variable in the dataset. This is useful for adding a third dimension of information to the plot.

* **size:** This parameter allows you to vary the size of the points in the plot based on a variable in the dataset, adding yet another dimension of information.

```python
# relplot-> figure level -> square shape
sns.relplot(data=d1,x="total_bill",y="tip",kind="scatter",hue="smoker",size="size")
```

**style:** You can use the style parameter to differentiate the points using different markers or line styles based on a variable in the dataset.

```python
sns.relplot(data=d1,x="total_bill",y="tip",hue="sex",style="time",size="size")
```

**Sizes:** The sizes parameter allows you to control the size of the points in a scatter plot based on a variable in your dataset

```python
sns.relplot(data=d1,x="total_bill",y="tip",hue="sex",style="time",size="size",sizes=(15,200))
```

# Lineplot

```python
d2=pd.read_csv(r"D:\Python\Datasets\gapminder_data_graphs.csv")
d2
```

```python
d2_india=d2[d2["country"]=="India"]
d2_india
```

```python
# Axis level
sns.lineplot(data=d2_india,x="year",y="life_exp")
```

```python
# figure level
sns.relplot(data=d2_india,x="year",y="life_exp",kind="line")
```

```python
d2_c=d2[d2["country"].isin(["India","Pakistan","China"])]
d2_c
```

```python
sns.relplot(data=d2_c,x="year",y="life_exp",kind="line",hue="country")
```

```python
d3_d=d2[d2["country"].isin(["India","Nigeria","Brazil"])]
d3_d
```

```python
sns.relplot(data=d3_d,x="year",y="life_exp",hue="country",style="continent",kind="line")
```

```python
sns.relplot(data=d3_d,x="year",y="life_exp",hue="country",style="continent",kind="line",size="continent")
```

# Facet Plot - Same plots as subplots

Facet plot will only work with figure level plots.

```python
sns.relplot(data=d1,x="total_bill",y="tip",kind="scatter",col="sex")
```

```python
sns.relplot(data=d1,x="total_bill",y="tip",kind="scatter",row="smoker")
```

```python
sns.relplot(data=d1,x="total_bill",y="tip",kind="scatter",col="sex",row="day")
```

```python
sns.relplot(data=d2,x="life_exp",y="gdp",kind="scatter",col="year",col_wrap=3)
```
