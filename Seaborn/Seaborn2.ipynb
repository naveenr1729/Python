```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2) Distribution Plots
* Used for univariate analysis.
* Used to find out the distribution.
* Range of the observation.
* Central Tendency.
* Is the data bimodal.
* Are there bimodal.

* figure level-displot.
* axis level plots-histplot,kdeplot,rugplot.

```python
d1=sns.load_dataset("tips")
d1
```

```python
sns.histplot(data=d1,x="total_bill")
```

```python
sns.displot(data=d1,x="total_bill",bins=20)
```

```python
sns.displot(data=d1,x="day",shrink=0.8)
```

```python
sns.displot(data=d1,x="tip",hue="sex")
```

```python
sns.displot(data=d1,x="tip",hue="sex",element='step')
```

#   kde plot 
* Kernal Density Estimation plot.
* Rather than using discrete bins, the kde plot smooths the observations with a Gaussian Kernal (a function), producing a continuous estimate.

```python
sns.kdeplot(data=d1,x="total_bill")
```

```python
sns.displot(data=d1,x="total_bill",kind="kde",hue="sex")
```

```python
sns.displot(data=d1,x="total_bill",kind="kde",hue="sex",fill=True)
```

# To increase the size

```python
sns.displot(data=d1,x="total_bill",kind="kde",hue="sex",fill=True,height=10,aspect=2)
```

# Rugplot
* Plots marginal distributions by drawing ticks along the x and y axis.
* This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.kdeplot(data=d1,x="total_bill")
sns.rugplot(data=d1,x="total_bill")
```

# Bivariate histogram

```python
sns.displot(data=d1,x="total_bill",y="tip")
```

```python
# Bivariate kdeplot
# A bivariate KDE plot smooths the (x,y) observations with a 2D Gaussian.
sns.displot(data=d1,x="total_bill",y="tip",kind="kde")
```
