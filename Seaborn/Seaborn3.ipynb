<h1 align="center">Matrix Plots Using Seaborn</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as px
```

# Heat Map


```python
d1=pd.read_csv("D:\Python\Datasets\gapminder_data_graphs.csv")
d1
```

```python
d1["country"].unique().sum()
```

```python
temp_d1=d1.pivot(index="country",columns="year",values="life_exp")
temp_d1
```

```python
d1["country"].nunique()
```

```python
#axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_d1)
```

```python
temp1_d1=d1[d1["continent"]=="Europe"].pivot(index="country",columns="year",values="life_exp")
temp1_d1
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp1_d1,annot=True)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp1_d1,annot=True,linewidths=0.4,cmap="viridis")
```
